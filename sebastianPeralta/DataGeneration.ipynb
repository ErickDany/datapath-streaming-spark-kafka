```python
import pandas as pd
import numpy as np
import random
import json
```

```python
matches = []
teams = np.arange(1,21)

while len(teams) != 0:
    one = random.randint(1,20)
    two = random.randint(1,20)
    if one in teams and two in teams and one != two:
        matches.append([one,two])
        teams = np.setdiff1d(teams,matches[len(matches)-1])
        print(teams)

print(matches)
```

```python
def get_data(matchday, min):
    df = pd.DataFrame()
    size = 10
    df['matchday'] = np.full((size,), matchday) #np.random.randint(0,100, size)
    df['minute'] = np.full((size,), min)
    df['date'] = np.full((size,), "2022-03-27")
    df['timestamp'] = np.full((size,), "2022-03-27 10:05:00")
    
    df['teamId_1'] = [x[0] for x in matches]
    df['teamId_2'] = [x[1] for x in matches]
    
    df['goal_1'] = np.full((size,), 0)
    df['goal_2'] = np.full((size,), 0)
    
    df['poss_1'] = np.random.randint(30, 70, size)
    df['poss_2'] = 100-df['poss_1']
    
    df['passes_attp_1'] = np.random.randint(5, 40, size)
    df['pass_comp_1'] = [x/10 for x in random.sample(range(400, 1000), size)]
    
    df['shots_attp_1'] = np.random.choice([0,0,0,0,1,1,2,3,4], size)
    df['shots_trg_1'] = df['shots_attp_1']*(df['pass_comp_1']/100).astype('int')
    
    df['passes_attp_2'] = np.random.randint(5, 40, size)
    df['pass_comp_2'] = [x/10 for x in random.sample(range(0, 1000), size)]
    
    df['shots_attp_2'] = np.random.choice([0,0,0,0,1,1,2,3,4], size)
    df['shots_trg_2'] = df['shots_attp_2']*(df['pass_comp_2']/100).astype('int')

    return df
```
